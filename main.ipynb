Edit

2. Imputation (Filling Missing Data)
ðŸ‘‰ Instead of removing, you fill the missing data with some meaningful value.

Options for Imputation:

Fill with Mean (for numeric data):

python
Copy
Edit
df['ColumnName'].fillna(df['ColumnName'].mean(), inplace=True)
Fill with Median (good for skewed numeric data):

python
Copy
Edit
df['ColumnName'].fillna(df['ColumnName'].median(), inplace=True)
Fill with Mode (for categorical data):

python
Copy
Edit
df['ColumnName'].fillna(df['ColumnName'].mode()[0], inplace=True)
Fill with a Constant Value:

python
Copy
Edit
df['ColumnName'].fillna(0, inplace=True)
# or for categorical:
df['ColumnName'].fillna('Unknown', inplace=True)
Forward Fill (ffill) or Backward Fill (bfill): Useful in time series data.

python
Copy
Edit
df.fillna(method='ffill', inplace=True)
df.fillna(method='bfill', inplace=True)

## ðŸ”· LotFrontage (259 missing)
##Type: Numerical (continuous)

##Meaning: Linear feet of street connected to property.

train_data['LotFrontage'].fillna(train_data['LotFrontage'].median(), inplace=True)



```python
train_data['LotFrontage'] = train_data['LotFrontage'].fillna(train_data['LotFrontage'].median())

train_data['LotFrontage'].isnull().sum()
```

##ðŸ”· MasVnrArea (8 missing)


##Meaning: Masonry veneer area in square feet.

##Best Fix: Fill with 0 (if MasVnrType is also missing, it's likely there's no veneer).

```python
train_data['MasVnrArea'].fillna(0, inplace=True)
```

```python
train_data['GarageType']=train_data['GarageType'].fillna(0)
```

 Basement-related columns:
These often have missing values because the house has no basement, so:

BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1, BsmtFinType2

Type: Categorical

Fix: Fill with "None" to indicate no basement.

```python
bsmt_cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
for col in bsmt_cols:
    train_data[col].fillna('None', inplace=True)
```

 Electrical (1 missing)
Type: Categorical

Best Fix: Fill with mode (most common electrical system).

```python
train_data['Electrical'].fillna(train_data['Electrical'].mode()[0], inplace=True)
```

Garage-related columns:
These often have missing values because there is no garage.

GarageType, GarageYrBlt, GarageFinish, GarageQual, GarageCond

Fix:

For categorical columns â†’ Fill with "None"

For GarageYrBlt (numerical) â†’ Fill with 0



```python
garage_cat_cols = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
for col in garage_cat_cols:
    train_data[col].fillna('None', inplace=True)

train_data['GarageYrBlt'].fillna(0, inplace=True)
```
